```python
# importing the required module
import matplotlib.pyplot as plt
  
```

```python
#Example of long 
stock_value = 10
purchase_volume = 100
increase_increment = .03
range_months = range(1,12)
ser = []


for new_price in range_months:
    purchase_volume += (purchase_volume * increase_increment)
    ser.append(purchase_volume)
print(ser)

plt.plot(range_months, ser)
plt.xlabel('months')
plt.ylabel('value')
plt.title('example of successful long')
plt.show()
```

```python
#example short
borrow_shares = 100
share_price = 10
sell_value = borrow_shares * share_price
share_change = [x for x in range( share_price, 1, -1)]
print(share_change)
investor_value = [sell_value - (borrow_shares * value) for value in share_change]
print(investor_value)
```

```python
plt.plot(share_change, investor_value)
ax = plt.gca()
ax.invert_xaxis()
plt.xlabel('share price')
plt.ylabel('value')
plt.title('example of successful short')
plt.show()
```
